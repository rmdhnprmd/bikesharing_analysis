## Exploratory Data Analysis

### Explore

````python
day_correlation = day_df.select_dtypes(include='number').corr()
day_correlation
````

````python
plt.figure(figsize=(12, 8))
sns.heatmap(day_correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasi Heatmap dari day_df')
plt.show()
````

**Insight:**
- korelasi antara ```registered``` dan ```cnt``` (**0.95**) menunjukkan sebagian besar total penyewaan sepeda berasal dari pengguna terdaftar
- korelasi antara ```casual``` dan ```cnt``` (**0.67**) menunjukkan bahwa pengguna ```casual``` juga memberikan kontribusi signifikan
- ```season``` dan ```cnt``` (**0.41**) ada korelasi positif moderat antara musim dan jumlah penyewaan sepeda. Ini menunjukkan bahwa penyewaan sepeda lebih banyak terjadi di musim-musim tertentu, meskipun korelasi ini tidak sangat kuat
- ```temp``` dan ```atemp``` berhubungan kuat dengan ```cnt```, suhu normal (```temp```) dan suhu yang dirasakan oleh manusia (```atemp```) memiliki korelasi positif dengan penyewaan sepeda sekitar (**0.63**)

````python
day_df[['casual', 'registered']].mean().plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Average Casual vs Registered Bike Rentals')
plt.ylabel('Average Rentals')
plt.xticks(rotation=0)
plt.show()
````

````python
day_season_insight = day_df.groupby('season')['cnt'].mean().reset_index()
day_season_insight['season'] = day_season_insight['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_season_insight['cnt'] = day_season_insight['cnt'].astype(int)
day_season_insight
````

````python
plt.figure(figsize=(8, 6))
colors_ = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3",]

sns.barplot(
    x='cnt',
    y='season',
    data=day_season_insight,
    palette=colors_,
    hue='season',)

# Menambahkan judul dan label pada sumbu
plt.title('Rata-rata Penyewaan Sepeda per Season day_df', loc='center', fontsize=14)
plt.xlabel('Average', fontsize=12)
plt.ylabel('Season', fontsize=12)

plt.show()
````

**Insight:**
- Musim gugur (```Fall```) memiliki rata-rata penyewaan sepeda tertinggi, yaitu sekitar **5644**
- Musim panas (```Winter```) dan musim dingin (```Summer```) memiliki rata-rata yang cukup mirip, dengan sekitar **4728** hingga **4992**
- Musim semi (```Spring```) memiliki rata-rata penyewaan sepeda yang lebih rendah, sekitar **2604**

````python
# Rata-rata penyewaan berdasarkan workingday
workingday_insight = day_df.groupby('workingday')['cnt'].mean().reset_index()

# Rata-rata penyewaan berdasarkan weekday
weekday_insight = day_df.groupby('weekday')['cnt'].mean().reset_index()

# Rata-rata penyewaan berdasarkan holiday
holiday_insight = day_df.groupby('holiday')['cnt'].mean().reset_index()

# Visualisasi: Workingday
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.barplot(x='workingday', y='cnt', data=workingday_insight, palette='viridis')
plt.title('Rata-rata Penyewaan (Workingday)')
plt.xlabel('Workingday (0: Bukan, 1: Ya)')
plt.ylabel('Rata-rata Penyewaan')

# Visualisasi: Weekday
plt.subplot(1, 3, 2)
sns.barplot(x='weekday', y='cnt', data=weekday_insight, palette='coolwarm')
plt.title('Rata-rata Penyewaan (Weekday)')
plt.xlabel('Weekday (0: Minggu, 1: Senin, dst.)')
plt.ylabel('Rata-rata Penyewaan')

# Visualisasi: Holiday
plt.subplot(1, 3, 3)
sns.barplot(x='holiday', y='cnt', data=holiday_insight, palette='mako')
plt.title('Rata-rata Penyewaan (Holiday)')
plt.xlabel('Holiday (0: Bukan, 1: Ya)')
plt.ylabel('Rata-rata Penyewaan')

# Menampilkan plot
plt.tight_layout()
plt.show()
````
